```python
# จากคุณ Clement Twumasi และ Lutz Lehmann 
import numpy as np
import matplotlib.pyplot as plt
#from scipy.integrate import odeint


def fx(x,y,z,t): return sigma*(y-x)
def fy(x,y,z,t): return x*(rho-z)-y
def fz(x,y,z,t): return x*y-beta*z

#a) Defining the classical Runge-Kutta 4th order method

def RungeKutta4(x,y,z,fx,fy,fz,t,h):
    k1x,k1y,k1z = ( h*f(x,y,z,t) for f in (fx,fy,fz) ) 
    
    xs, ys,zs,ts = ( r+0.5*kr for r,kr in zip((x,y,z,t),(k1x,k1y,k1z,h)) )
    k2x,k2y,k2z = ( h*f(xs,ys,zs,ts) for f in (fx,fy,fz) )
    
    xs, ys,zs,ts = ( r+0.5*kr for r,kr in zip((x,y,z,t),(k2x,k2y,k2z,h)) )
    k3x,k3y,k3z = ( h*f(xs,ys,zs,ts) for f in (fx,fy,fz) )
    
    xs, ys,zs,ts = ( r+kr for r,kr in zip((x,y,z,t),(k3x,k3y,k3z,h)) )
    k4x,k4y,k4z  =( h*f(xs,ys,zs,ts) for f in (fx,fy,fz) )
    return (r+(k1r+2*k2r+2*k3r+k4r)/6 for r,k1r,k2r,k3r,k4r in 
            zip((x,y,z),(k1x,k1y,k1z),(k2x,k2y,k2z),(k3x,k3y,k3z),(k4x,k4y,k4z))) 
            # จากที่ศึกษาตรงนี้น่าจะเป็น y_{k+1} = y_k + \frac{\Delta{t}}{6}[f_1 + 2f_2 + 2f_3 + f_4]


sigma=10
beta=8/3
rho=28
tIn=0.
tFin=100
h=0.01
totalSteps=int(np.floor((tFin-tIn)/h))

t = totalSteps * [0]
x = totalSteps * [0]
y = totalSteps * [0]
z = totalSteps * [0]

x[0],y[0],z[0],t[0] = 1., 1., 1., 0.  #Initial condition
for i in range(1, totalSteps):
    x[i],y[i],z[i] = RungeKutta4(x[i-1],y[i-1],z[i-1], fx,fy,fz, t[i-1], h)

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')
ax.plot(x,y,z,'r',label='Lorentz 3D Solution')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
plt.show()
```

```python
# อันนี้เป็นแบบที่เข้าใจง่ายขึ่้น
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def lorenz_equations(sigma, rho, beta, x, y, z):
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return dx, dy, dz

def runge_kutta_step(sigma, rho, beta, x, y, z, h):
    k1x, k1y, k1z = lorenz_equations(sigma, rho, beta, x, y, z)
    k2x, k2y, k2z = lorenz_equations(sigma, rho, beta, x + 0.5 * h * k1x, y + 0.5 * h * k1y, z + 0.5 * h * k1z)
    k3x, k3y, k3z = lorenz_equations(sigma, rho, beta, x + 0.5 * h * k2x, y + 0.5 * h * k2y, z + 0.5 * h * k2z)
    k4x, k4y, k4z = lorenz_equations(sigma, rho, beta, x + h * k3x, y + h * k3y, z + h * k3z)

    new_x = x + (h / 6.0) * (k1x + 2 * k2x + 2 * k3x + k4x)
    new_y = y + (h / 6.0) * (k1y + 2 * k2y + 2 * k3y + k4y)
    new_z = z + (h / 6.0) * (k1z + 2 * k2z + 2 * k3z + k4z)

    return new_x, new_y, new_z

def simulate_lorenz(sigma, rho, beta, x0, y0, z0, h, num_steps):
    x_vals, y_vals, z_vals = [x0], [y0], [z0]

    for _ in range(num_steps):
        x, y, z = runge_kutta_step(sigma, rho, beta, x_vals[-1], y_vals[-1], z_vals[-1], h)
        x_vals.append(x)
        y_vals.append(y)
        z_vals.append(z)

    return x_vals, y_vals, z_vals

# Parameters for the Lorenz system
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Initial conditions
x0, y0, z0 = 1, 1, 1

# Time step and number of steps
h = 0.01
num_steps = 10000

# Run simulation
x_vals, y_vals, z_vals = simulate_lorenz(sigma, rho, beta, x0, y0, z0, h, num_steps)

# Plot the results
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_vals, y_vals, z_vals, lw=0.5)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Lorenz Attractor')

plt.show()
```
